```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
from white_brush.io import read_image
from white_brush.color_balance import balance_color
```

```python
def side_by_side(image_name : str):
    img = read_image(image_name)
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(img)
    ax1.axis("off")
    ax2.imshow(balance_color(img))
    ax2.axis("off")
    plt.gcf().set_size_inches(15, 8)
    print(image_name[3:] if image_name.startswith("../") else image_name)
```
